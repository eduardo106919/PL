# Context-free grammars

Context-free grammars (CFGs) are used to specify context-free languages. A CFG consists of:
  - a set $\Sigma$ of *terminal symbols* (the alphabet of the language)
  - a set $N$ of *nonterminal symbols* (variables of the grammar)
  - *productions* that expand a nonterminal into a sequence of terminals and nonterminals
  - a *start symbol* $S_0$, selected from the nonterminals

A word $w \in \Sigma^*$ is derivable by a grammar if there is a *derivation* or a *parse tree*.
